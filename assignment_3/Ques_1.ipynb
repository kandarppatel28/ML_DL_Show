```python
#importing lib
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
results = {}
```

```python
#creating array of x and y such that y=(x/2) - 4
X = np.array([], dtype=float)
Y = np.array([], dtype=float)
for i in range(0, 11, 2):
    y = ( i / 2) - 4
    X = np.append(X,i)
    Y = np.append(Y,y)
print(X)
print(Y)
```

```python
#Shape[1], Optimiser: sgd, loss: mean_squared_error, Epochs: 200
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(X, Y, epochs=200)
print(model.predict([40.0]))
results['Shape[1], Optimiser: sgd, loss: mean_squared_error, Epochs: 200'] = (model.predict([40.0]))
```
